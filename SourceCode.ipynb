```python
# finding the 1st quartile
q1 = np.quantile(df['Radius'], 0.25)
 
# finding the 3rd quartile
q3 = np.quantile(df['Radius'], 0.75)
med = np.median(df['Radius'])
 
# finding the interquartile region (ipr)
iqr = q3-q1
 
# finding upper and lower whiskers
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print('iqr:         ', iqr, '\nupper_bound: ', upper_bound, '\nlower_bound: ', lower_bound)
```

```python
outliers = df['Radius'][(df['Radius'] <= lower_bound) | (df['Radius'] >= upper_bound)]
print('Outlier Count: ', outliers.count())
print('The following are the outliers in the boxplot:\n {} '.format(outliers))
```

```python
x = df.copy()
x.drop(['Diagnosis'], axis=1, inplace=True)
y = df.copy()
y = y['Diagnosis']
```

```python
# Splitting Data between train & test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=33)
```

```python
xtc = x_train.count()
ytc = y_train.count()
xpc = x_test.count()
ypc = y_test.count()
print('Training Data:', '\nDiagnosis           ', ytc, '\n', xtc)
print('\nTesting Data:', '\nDiagnosis           ', ypc, '\n', xpc)
```

```python
# Importing StandardScaler
from sklearn.preprocessing import StandardScaler
# Creating object
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# Making Keras Model
from keras.models import Sequential
from keras.layers import Dense
# Declaring model
model = Sequential()
# 1st hidden layer
model.add(Dense(20, input_dim=10, activation='relu'))
# 2nd hidden layer
model.add(Dense(10, activation='relu'))
# Last/output layer
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=100, batch_size=10)
_, accuracy = model.evaluate(x_train, y_train)
print("Accuracy: ", accuracy)
```

```python
#saving the model
model.save('Breast Cancer Model.h5')
```

```python
#now testing for Test data
y_pred = model.predict(x_test)
```

```python
#converting values
y_pred = (y_pred>0.5)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test,y_pred)
score = accuracy_score(y_test,y_pred)
print(cm)
print('Test score is:',score)
```

```python
# Creating Heatmap of Confussion matrix
plt.figure(figsize=[14,7])
sb.heatmap(cm, annot=True, fmt='')
plt.ylabel("Actual", fontsize='15')
plt.xlabel("Predicted", fontsize='15')
plt.show()
```
